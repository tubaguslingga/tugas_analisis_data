## Visualization & Explanatory Analysis

### Pertanyaan 1: Produk apa yang memiliki tingkat penjualan tertinggi dan terendah selama kuartal terakhir?


```python
sum_order_df = all_data.groupby("product_category_name_english")["product_id"].count().reset_index()
sum_order_df = sum_order_df.rename(columns={"product_id": "products"})
sum_order_df = sum_order_df.sort_values(by="products", ascending=False)
sum_order_df = sum_order_df.head(10)

sum_order_df.head()
```

```python
figr, a_x = mplt.subplots(nrows=1, ncols=2, figsize=(24, 6))

colors = ["blue", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sbr.barplot(x="products", y="product_category_name_english", data=sum_order_df.head(5), palette=colors, ax=a_x[0], hue="product_category_name_english", legend=False)
a_x[0].set_ylabel(None)
a_x[0].set_xlabel(None)
a_x[0].set_title("Produk paling tidak laku", loc="center", fontsize=18)
a_x[0].tick_params(axis='y', labelsize=15)

sbr.barplot(x="products", y="product_category_name_english", data=sum_order_df.sort_values(by="products", ascending=True).head(5), palette=colors, ax=a_x[1], hue="product_category_name_english", legend=False)
a_x[1].set_ylabel(None)
a_x[1].set_xlabel(None)
a_x[1].invert_xaxis()
a_x[1].yaxis.set_label_position("right")
a_x[1].yaxis.tick_right()
a_x[1].set_title("Produk paling laku", loc="center", fontsize=18)
a_x[1].tick_params(axis='y', labelsize=15)

mplt.suptitle("Produk paling laku dan paling tidak laku", fontsize=20)
mplt.show()
```

### Pertanyaan 2: Berapa rata-rata pengeluaran pelanggan dalam beberpa bulan terakhir?


```python
mntly_spend_df = all_data.resample(rule='ME', on='order_approved_at').agg({
    "payment_value":"sum"
})
mntly_spend_df.index = mntly_spend_df.index.strftime('%B') #mengubah format order_approved_at menjadi Tahun-Bulan
mntly_spend_df = mntly_spend_df.reset_index()
mntly_spend_df.rename(columns={
    "payment_value":"total_spend"
}, inplace=True)
mntly_spend_df.head()
```

```python
mntly_spend_df = mntly_spend_df.sort_values('total_spend').drop_duplicates('order_approved_at', keep='last')
```

```python
mntly_spend_df.head()
```

```python
mntly_spend_df.sort_values(by='total_spend')
```

```python
month_mapping = {
    "January": 1,
    "February": 2,
    "March": 3,
    "April": 4,
    "May": 5,
    "June": 6,
    "July": 7,
    "August": 8,
    "September": 9,
    "October": 10,
    "November": 11,
    "December": 12
}
mntly_spend_df["month_numeric"] = mntly_spend_df["order_approved_at"].map(month_mapping)
mntly_spend_df = mntly_spend_df.sort_values("month_numeric")
mntly_spend_df = mntly_spend_df.drop("month_numeric", axis=1)
```

```python
mplt.figure(figsize=(10, 5))
mplt.plot(
    mntly_spend_df["order_approved_at"],
    mntly_spend_df["total_spend"],
    marker='o',
    linewidth=2,
    color="blue"
)
mplt.title("Total uang yang dibelanjakan pelanggan per Bulan (2018)", loc="center", fontsize=20)
mplt.xticks(fontsize=10, rotation=25)
mplt.yticks(fontsize=10)
mplt.show()
```

### Pertanyaan 3: Berapa tingkat kepuasan dari pembeli terhadap layanan kami?


```python
review_scores = all_data['review_score'].value_counts().sort_values(ascending=False)

most_common_score = review_scores.idxmax()

sbr.set(style="darkgrid")

mplt.figure(figsize=(10, 5))
sbr.barplot(x=review_scores.index,
            y=review_scores.values,
            order=review_scores.index,
            color="blue"  
            )

mplt.title("Peringkat oleh pelanggan untuk service", fontsize=15)
mplt.xlabel("Peringkat")
mplt.ylabel("Count")
mplt.xticks(fontsize=12)
mplt.legend([], frameon=False) 
mplt.show()
```

### Pertanyaan 4: Wilayah geografis mana yang menghasilkan jumlah pelanggan terbanyak?
